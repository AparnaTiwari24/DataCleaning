```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Aparna Tiwari\Downloads\Dataset 1_origin.csv")
```

```python
# Drop constant-value columns
# These are columns with only one unique value (no useful information)
for col in df.columns:
    if df[col].nunique() == 1:
        df.drop(columns=col, inplace=True)
```

```python
#Drop duplicate columns
# 'marital_statues' is likely a typo or duplicate of 'marital_status'
if 'marital_statues' in df.columns:
    df.drop(columns='marital_statues', inplace=True)
```

```python
#Convert 'dt_customer' to datetime format
df['dt_customer'] = pd.to_datetime(df['dt_customer'], errors='coerce')
```

```python
#Rename all columns to lowercase and snake_case
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
#Standardize text fields (remove extra spaces, proper case)
text_cols = df.select_dtypes(include='object').columns
for col in text_cols:
    df[col] = df[col].astype(str).str.strip().str.title()
```

```python
#Convert year_birth to age and drop year_birth
df['age'] = 2025 - df['year_birth']
df.drop(columns='year_birth', inplace=True)
```

```python
#Handle missing values (you can customize this)
# Option 1: Drop rows with missing values
df = df.dropna()
```

```python
# Fill specific columns (e.g., income) with median
df['income'] = df['income'].fillna(df['income'].median())
```

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

```python
#Save the cleaned data
df.to_csv(r"C:\Users\Aparna Tiwari\Downloads\Dataset 1_origin.csv", index=False)
```
